```python
DF= pd.merge(DF,Hourly_AVG, on=['Weekdays/Weekends','Hour(24-Pac)'], how='left')
DF= pd.merge(DF,Hourly_STD, on=['Weekdays/Weekends','Hour(24-Pac)'], how='left')
DF= pd.merge(DF,Hourly_Count, on=['Weekdays/Weekends','Hour(24-Pac)'], how='left')
```

```python
DF.loc[DF['Weekdays/Weekends']=='Weekday', 
       'Wk/Wknd Upper CI(95%)']= DF['Hourly AVG']+(2.08* (DF['Hourly STD']/(DF['Hourly Count']**(1/2))))
DF.loc[DF['Weekdays/Weekends']=='Weekday',
       'Wk/Wknd Lower CI(95%)']= DF['Hourly AVG']-(2.08* (DF['Hourly STD']/(DF['Hourly Count']**(1/2))))

DF.loc[DF['Weekdays/Weekends']=='Weekend', 
       'Wk/Wknd Upper CI(95%)']= DF['Hourly AVG']+(2.447* (DF['Hourly STD']/(DF['Hourly Count']**(1/2))))
DF.loc[DF['Weekdays/Weekends']=='Weekend',
       'Wk/Wknd Lower CI(95%)']= DF['Hourly AVG']-(2.447* (DF['Hourly STD']/(DF['Hourly Count']**(1/2))))

```

```python
Con1=[
    (DF['Wk/Wknd Above/Below']== 'Above'), (DF['Wk/Wknd Above/Below']== 'Below'), (DF['Wk/Wknd Above/Below']== 'Neutral'),
]

Values1=[(DF['Minutes per Minute']-DF['Wk/Wknd Upper CI(95%)']),
         (DF['Minutes per Minute']-DF['Wk/Wknd Lower CI(95%)']),
         0]
DF['Above/Below (Dist)']= np.select(Con1,Values1, default=np.nan)
```
